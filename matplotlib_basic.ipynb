```python
# THƯ VIỆN VẼ BIỂU ĐỒ : matplotlib
'''
matplotlib là thư viện dùng để vẽ biểu đồ như:

  + Biểu đồ đường (line plot)

  + Biểu đồ cột (bar chart)

  + Biểu đồ phân tán (scatter plot)

  + Biểu đồ tròn (pie chart)

Và nhiều loại khác.
'''
# 1. Vẽ biểu đồ đường: (line plot)
# DÙNG CHO DỮ LIỆU LIÊN TỤC THEO THỜI GIAN
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.title("Biểu đồ đường: ")
plt.xlabel("x")
plt.ylabel("y")
# plt.scatter(x, y)
plt.show()

```

```python
# 2. Biểu đồ cột (bar chart)
# DÙNG ĐỂ SO SÁNH GIỮA CÁC NHÓM

labels = ['A', 'B', 'C', 'D', 'E']
values = [10, 5, 8, 7, 14]

plt.bar(labels, values)
plt.title("Biểu đồ cột: ")
plt.xlabel("Nhãn")
plt.ylabel("Giá trị")
plt.show()
```

```python
# 3. Biểu đồ phân tán (scatter)
# DÙNG ĐỂ QUAN SÁT MỐI QUAN HỆ GIỮA HAI BIẾN
x = [1, 2, 3, 4, 5, 3, 2, 4]
y = [2, 1, 4, 3, 0, 4, 5, 7]

plt.scatter(x, y)
# plt.plot(x, y)
plt.title("Biểu đồ phân tán")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# 4. Biểu đồ tròn (Pie Chart)
# DÙNG ĐỂ THỂ HIỆN TỈ LỆ PHẦN TRĂM
labels = ["A", "B", "C", "D"]
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Biểu đồ tròn")
plt.show()
```

```python
# 5. Biểu đồ tần suất
# DÙNG ĐỂ XEM PHÂN BỐ DỮ LIỆU
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Histogram - Phân bố dữ liệu")
plt.show()
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

model = tree.DecisionTreeClassifier()
# feature: [1, 2] và [3, 4]
# label: [0, 1]
model.fit([[1,2],[3,4]], [0, 1])

plt.figure(figsize=(8, 6))# <-- Tăng kích thước
tree.plot_tree(
    model,
    filled=True,  # Tô màu nút
    rounded=True, # Bo góc nút
    feature_names=["X1", "X2"],   #Đặt tên cột
    class_names=["Không", "Có"],  # Đặt tên nhãn
    fontsize=12  # <-- Tăng cỡ chữ
)
plt.tight_layout()  # Giúp căn lề
plt.show()
```
